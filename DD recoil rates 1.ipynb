```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f_mb(v, v_char):
    """Maxwell-Boltzmann velocity distribution.
    """
    t_over_m = v_char**2 / 3.
    
    return (1. / (2*np.pi*t_over_m)**1.5 * 4*np.pi*v**2 *
            np.exp(-v**2 / (2.*t_over_m)))
```

```python
v_mw = 1e-3
vs = np.logspace(-7, -2, 200)

plt.loglog(vs, f_mb(vs, v_mw) + f_mb(vs, 1e-5))
```
