```python
import pickle as pk
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('../../nfw_gamma_0.5/ds.pkl') as fh:
    ds = pk.load(fh)
with open('../../nfw_gamma_0.5/rss.pkl') as fh:
    rss = pk.load(fh)
```

```python
likelihoods = np.loadtxt('nfw_logmmin_5.csv', delimiter=',')
X, Y = np.meshgrid(ds, rss)
plt.contour(X, Y, np.log10(likelihoods))
```

```python
d, rs, p = np.loadtxt('nfw_gamma_0.5.csv', delimiter=',').T
d = d.reshape(len(np.unique(d)), -1)
rs = rs.reshape(d.shape)
p = p.reshape(d.shape)
plt.contour(d, rs, np.log10(p))
plt.xscale('log')
plt.yscale('log')
```
